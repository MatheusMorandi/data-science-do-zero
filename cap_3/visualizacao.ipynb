## Capítulo 3: Visualização de Dados

Neste capítulo, Joel Grus introduz a importância da visualização de dados como uma ferramenta fundamental para entender e comunicar insights extraídos de conjuntos de dados. A visualização é essencial para a análise exploratória de dados, ajudando a identificar padrões, tendências, outliers, e outros aspectos relevantes.

### Principais Tópicos Abordados

- **Matplotlib**: O capítulo apresenta a biblioteca `matplotlib`, uma das ferramentas mais populares para criar visualizações em Python. A partir de exemplos básicos, o leitor é ensinado a gerar gráficos simples como linhas, barras, e dispersão (scatter plots).

- **Personalização de Gráficos**: Além de criar gráficos básicos, o capítulo também explora como personalizar visualizações para torná-las mais informativas e esteticamente agradáveis. Isso inclui modificar cores, rótulos, títulos, e a adição de legendas.

- **Histograma**: A criação de histogramas é detalhada, mostrando como eles são úteis para entender a distribuição de uma variável.

- **Plotagem de Várias Séries**: O autor mostra como plotar múltiplas séries de dados no mesmo gráfico, uma técnica útil para comparações diretas entre diferentes conjuntos de dados.

- **Subplots**: A técnica de criar subplots, ou múltiplos gráficos em uma única janela, também é abordada, permitindo a comparação visual de diferentes variáveis lado a lado.

### Exemplo de Código

Ao longo do capítulo, são fornecidos exemplos práticos de como usar `matplotlib` para criar gráficos, personalizar visualizações, e efetuar análises exploratórias básicas.


```python
from matplotlib import pyplot as plt

anos = [1950, 1960, 1970, 1980, 1990, 2000, 2010]

pib = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]

plt.plot(anos, pib, color='purple', marker='o', linestyle='solid')

plt.title("PIB Nominal")

plt.ylabel("Bilhões de $")

plt.show()
```
